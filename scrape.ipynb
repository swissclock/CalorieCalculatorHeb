```python
import requests
import json

url = "https://www.goleango.com/calculators/nutrition_calculator/utils/get_hebrew_nutritional_value_for_food_id_json.php"
nutrition_data = {}

# Iterate over a range of food_id values
for food_id in range(1, 2747):  # Adjust range as needed
    payload = {'food_id': food_id}
    response = requests.post(url, data=payload)
    
    if response.status_code == 200:
        try:
            data = response.json()
            if data:  # Ensure data is not empty
                nutrition_data[food_id] = data
                print(f"Data found for food_id {food_id}")
        except ValueError:
            print(f"No valid JSON for food_id {food_id}")
    else:
        print(f"No data for food_id {food_id} (status code: {response.status_code})")

# Save collected data to a JSON file
with open("nutrition_data.json", "w", encoding="utf-8") as file:
    json.dump(nutrition_data, file, ensure_ascii=False, indent=4)

print("Scraping complete. Data saved to nutrition_data.json.")
```

```python
url = "https://www.goleango.com/calculators/nutrition_calculator/utils/get_food_list_for_category_json.php"
all_foods = {}

# Iterate over a range of food_id values
for category in range(1, 28):  # Adjust range as needed
    payload = {'category': category}
    response = requests.post(url, data=payload)
    
    if response.status_code == 200:
        try:
            data = response.json()
            if data:  # Ensure data is not empty
                all_foods[category] = data
                print(f"Data found for category {category}")
        except ValueError:
            print(f"No valid JSON for category {category}")
    else:
        print(f"No data for category {category} (status code: {response.status_code})")

# Save collected data to a JSON file
with open("all_alphabetical.json", "w", encoding="utf-8") as file:
    json.dump(all_foods, file, ensure_ascii=False, indent=4)

print("Scraping complete. Data saved to all_alphabetical.json.")
```
